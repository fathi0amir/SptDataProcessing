```python
reload_modules()
# laf.holoviz_plot_msd_loglog_fast(df, bin_num=150)

# fig = laf.plotly_plot_norm_loglog_msd(df)
# fig.show(renderer="svg")

# laf.vega_plot_msd_loglog_fast(df, bin_size=0.03)

def plot_msd_vs_time_lag(df):
    hist, edges_x, edges_y = nlss.log_binned_histogram2d(df, x_column='Lag_T', y_column='MSD_norm', bins_per_decade=20)
# hist, edges_x, edges_y = nlss.lin_binned_histogram2d(df, x_column='Lag_T', y_column='MSD_norm', x_bin_width=0.02, y_bin_width=20)
    hist = hist.astype(float)  # Convert to float to allow NaN values
    hist[hist <= 1] = np.nan  # Replace zeros with NaN for better visualization
    hv.renderer("bokeh").theme = Theme(filename="util/bokeh-theme-light.yaml")
    msd_quad = hv.QuadMesh((edges_x, edges_y, hist.T)).opts(
    logx=True, 
    logy=True, 
    xlabel='Lag Time (s)', 
    ylabel='MSD / MSD(Δt=0)', 
    title='Heatmap of MSD vs Lag Time', 
    xlim=(1e-2, 1e1),
    ylim=(1e0, 1e3),
    # colorbar=True,
    cmap='Reds',
    clim=(0, 2500))
    msd_quad.opts(laf.HV_BOKEH_BASIC)
    msd_quad.opts(hooks=[laf.bokeh_add_topright_log_axes])
    return msd_quad

plot_msd_vs_time_lag(df)
```

### Find confinement ratio for each group

```python
def calculate_confinement_ratio_traj_group():
    # One row per trajectory
    traj_level = df.groupby('UID', as_index=False).first()

# Confinement ratio = fraction of sub-diffusive trajectories in each traj_label group
    confinement_ratio = (
    traj_level.assign(is_confined=traj_level['Alpha_Flag_Fit'].eq('sub'))
    .groupby('traj_label')['is_confined']
    .agg(['sum', 'count'])
    .assign(confinement_ratio=lambda x: x['sum'] / x['count'])
    .reset_index()
    .rename(columns={'sum': 'n_confined', 'count': 'n_total'})
)

    return confinement_ratio

calculate_confinement_ratio_traj_group()
```

### Find confinement domains (size and number) for each group

```python
# 1. Calculate the length of each trajectory (number of steps)
# We can count the number of rows for each unique UID
def get_long_trajectories(df, traj_length_threshold=300, traj_label='cross'):
    trajectory_lengths = df.groupby('UID')['Frame'].count().reset_index()
    trajectory_lengths.rename(columns={'Frame': 'Trajectory_Length'}, inplace=True)

# 2. Identify UIDs that have a length greater than the threshold
    long_trajectories_uids = trajectory_lengths[trajectory_lengths['Trajectory_Length'] > traj_length_threshold]['UID'].unique()

# 3. Filter the original df for 'traj_label == 'in'' and for the identified UIDs
    df_in_long_trajectories = df[
        (df['traj_label'] == traj_label) & 
        (df['UID'].isin(long_trajectories_uids))
    ]   
    
    return df_in_long_trajectories

df_in_long_trajectories = get_long_trajectories(df, traj_length_threshold=500, traj_label='out')

print(f"Number of trajectories in df_in_long_trajectories: {len(df_in_long_trajectories['UID'].unique())}")
df_in_long_trajectories
```
