### Load data

```python
reload_modules()
data_wd = Path(r"data\2 50 TAT\BMP50TAT-20260610-S2-00267")
# df = dpp.load_trackmate_csv_files(data_wd)
df = dpp.load_trackpy_parquet(data_wd, suffix='-tracks', exposure_time=const.DT, pixel_in_nm=const.PIXEL_IN_NM)
print(f"Number of tracks before filtering: {len(df['UID'].unique())}")
df
```

### Remove Stationary and short Tracks

```python
reload_modules()
df = nlss.drop_short_trajectories(df)
print(f"Number of tracks after filtering short trajectories: {len(df['UID'].unique())}")
df = nlss.drop_stationary_trajectories(df)
print(f"Number of tracks after filtering stationary trajectories: {len(df['UID'].unique())}")
```

### Calculate MSD

```python
reload_modules()
df = df.groupby('UID').apply(nlss.calculate_msd).reset_index(level='UID')
```

### Normalized MSD (To first point)

```python
reload_modules()
df['MSD_norm'] = df.groupby('UID')['MSD'].transform(lambda x: x / x.iloc[0])
```

### Calculate Normal Diffusion Coefficient

```python
reload_modules()
df = df.groupby('UID').apply(nlss.calculate_diff_d)
df = df.reset_index(level='UID') if 'UID' not in df.columns else df.reset_index(drop=True)
```

### Flagging alpha values based on linear fit (MSD vs time lag) and find the corresponding D values
Alpha values are grouped into 3 categories:
- In: alpha $\leq$ 0.8
- Cross: 0.8 $\leq$ alpha $\leq$ 1.2
- Out: alpha $\geq$ 1.2

Values are ignored if $\leq$ 0.2. Criteria are set in `util/constants.py` file.

```python
df = df.groupby('UID').apply(nlss.flag_alpha_by_fit)
df = df.reset_index(level='UID') if 'UID' not in df.columns else df.reset_index(drop=True)

df = df.groupby('UID').apply(nlss.calc_d_fix_alpha)
df = df.reset_index(level='UID') if 'UID' not in df.columns else df.reset_index(drop=True)
```

### Group by in/out/cross
*traj_label* column should be present in the dataframe.

```python
df_in = df[df['traj_label'] == 'in']
df_out = df[df['traj_label'] == 'out']
df_cross = df[df['traj_label'] == 'cross']
```

### Plot Diffusion Coefficient Histogram

```python
reload_modules()

hist, edges = nlss.log_binned_histogram(df_cross, column='D_Fixed_Alpha', min_log_d=-3, max_log_d=0, bins_per_decade=10)

hv.renderer("bokeh").theme = Theme(filename="util/bokeh-theme-light.yaml")
d_hist = hv.Histogram((hist, edges)).opts(logx=True, logy=False, xlabel='D (µm²/s)', ylabel='Count', title='Histogram of D')
d_hist.opts(laf.HV_BOKEH_BASIC)
d_hist.opts(hooks=[laf.bokeh_add_topright_logxliny_axes])
d_hist
```

### MSD vs Time Lag (Log-Log) Plot

```python
reload_modules()
# laf.holoviz_plot_msd_loglog_fast(df, bin_num=150)

# fig = laf.plotly_plot_norm_loglog_msd(df)
# fig.show(renderer="svg")

# laf.vega_plot_msd_loglog_fast(df, bin_size=0.03)

hist, edges_x, edges_y = nlss.log_binned_histogram2d(df, x_column='Lag_T', y_column='MSD_norm', bins_per_decade=20)
# hist, edges_x, edges_y = nlss.lin_binned_histogram2d(df, x_column='Lag_T', y_column='MSD_norm', x_bin_width=0.02, y_bin_width=20)
hist = hist.astype(float)  # Convert to float to allow NaN values
hist[hist < 5] = np.nan  # Replace zeros with NaN for better visualization
hv.renderer("bokeh").theme = Theme(filename="util/bokeh-theme-light.yaml")
msd_quad = hv.QuadMesh((edges_x, edges_y, hist.T)).opts(
    logx=True, 
    logy=True, 
    xlabel='Lag Time (s)', 
    ylabel='MSD / MSD(Δt=0)', 
    title='Heatmap of MSD vs Lag Time', 
    xlim=(1e-2, 1e1),
    ylim=(1e0, 1e3),
    # colorbar=True,
    cmap='viridis',
    clim=(0, 200))
msd_quad.opts(laf.HV_BOKEH_BASIC)
msd_quad.opts(hooks=[laf.bokeh_add_topright_log_axes])
msd_quad
```

### Plot Stats for in/out/cross groups

```python
reload_modules()
plot_df = nlss.domain_diff_type_stacked_df(df)

# 6. Create the stacked bars
bars = hv.Bars(
    plot_df,
    kdims=['traj_label', 'Alpha_Flag_Fit'],
    vdims=['Proportion']
).opts(
    stacked=True,
    xlabel='Trajectory Label',
    ylabel='Proportion',
    title='Diffusion Types by Trajectory Group',
    ylim=(0, 1.21),
    color='Alpha_Flag_Fit',
    show_legend=True,
    cmap={
        'normal': '#2ca02c',  # green
        'sub':    '#1f77b4',  # blue
        'sup':    '#ff7f0e',  # orange
    },
)

# 7. Create labels and overlay them on the bars
labels = hv.Labels(
    plot_df,
    kdims=['traj_label', 'label_y_pos'],
    vdims=['label']
).opts(
    text_color='white',
    text_align='center',
    text_baseline='middle'
)

# 8. Combine bars + labels and apply your theme
combined = bars * labels
combined.opts(laf.HV_BOKEH_BASIC)
combined.opts(hooks=[laf.bokeh_add_topright_linear_axes, laf.set_legend_top_horizontal])
combined
```
